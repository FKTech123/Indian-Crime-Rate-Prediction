# Implement machine learning model

```python
crime.columns
```

```python
min_val = crime['TOTAL IPC CRIMES'].min()
max_val = crime['TOTAL IPC CRIMES'].max()
range_val = (max_val - min_val) / 4
low = min_val + range_val
medium = low + range_val
high = medium + range_val
```

```python
def get_crime_level(crime_count):
    if crime_count <= low:
        return 1
    elif crime_count <= medium:
        return 2
    elif crime_count <= high:
        return 3
    else:
        return 4

crime['CRIME_LEVEL'] = crime['TOTAL IPC CRIMES'].apply(get_crime_level)
```

```python
crime_level_count = crime['CRIME_LEVEL'].value_counts()
crime_level_count
```

# Linear Regression

```python
X = crime.drop(['CRIME_LEVEL','STATE/UT', 'DISTRICT','TOTAL IPC CRIMES'], axis=1)
y = crime['TOTAL IPC CRIMES']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics
lr = LinearRegression()
lr.fit(X_train,y_train)
lr_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, precision_score, f1_score, recall_score, roc_curve, roc_auc_score
import matplotlib.pyplot as plt
```

```python
lr_pred = lr.predict(X_test)
lr_score = lr.score(X_test, y_test)
print('Linear Regression score : ',lr_score)
print()
print('Mean_absolute_error  = ',metrics.mean_absolute_error(lr_pred,y_test))
print()
print('Mean_squared_error   = ',metrics.mean_squared_error(lr_pred,y_test))
print()
print('R2_score             = ',metrics.r2_score(lr_pred,y_test))
print()
R2 = metrics.r2_score(lr_pred,y_test)
adj_R2 = 1-((1-R2)*(len(y)-1)/(len(y)-X.shape[1]-1))
print('Adjusted_R2         = ',adj_R2)
```

# Hierarchical _Clustering

```python
from scipy.cluster.hierarchy import linkage, dendrogram
Z = linkage(X_scaled, method='ward')

# Plot dendrogram
plt.figure(figsize=(20, 8))
dendrogram(Z)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Data point')
plt.ylabel('Distance')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

# Perform hierarchical clustering
hier_clust = AgglomerativeClustering(n_clusters=3)
labels = hier_clust.fit_predict(X_scaled)

# Visualize clusters
plt.scatter(crime['THEFT'], crime['MURDER'], c=labels)
plt.xlabel('Number of Thefts')
plt.ylabel('Number of Murders')
plt.show()
```

# Recurrent Neural Network

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
import math
import matplotlib.pyplot as plt
```

```python
#Initialize RNN:
RNN = Sequential()

#Adding the first RNN layer and some Dropout regularization
RNN.add(SimpleRNN(units = 50, activation='tanh', return_sequences=True, input_shape= (X_train.shape[1],1)))
RNN.add(Dropout(0.2))

#Adding the second RNN layer and some Dropout regularization
RNN.add(SimpleRNN(units = 50, activation='tanh', return_sequences=True))
RNN.add(Dropout(0.2))

#Adding the third RNN layer and some Dropout regularization
RNN.add(SimpleRNN(units = 50, activation='tanh', return_sequences=True))
RNN.add(Dropout(0.2))

#Adding the fourth RNN layer and some Dropout regularization
RNN.add(SimpleRNN(units = 50))
RNN.add(Dropout(0.2))

#Adding the output layer
RNN.add(Dense(units = 1))
```

```python
#Compile the RNN
RNN.compile(optimizer='adam', loss='mean_squared_error')

#Fitting the RNN to the Training set
RNN.fit(X_train, y_train, epochs=100, batch_size=32)
```

```python
# Make the Confusion Matrix
RNN_pred = RNN.predict(X_test)
RNN_pred = (RNN_pred > 0.5)
RNN_score = RNN.evaluate(X_test, RNN_pred)
print('Recurrent Neural Network score : ',RNN_score)
print()
RNN_cm = confusion_matrix(y_test, RNN_pred)
print('Confusion matrix: ')
print(RNN_cm)
print()
# print(classification_report(y_test, log_pred))
# accuracy: (tp + tn) / (p + n)
RNN_accuracy = accuracy_score(y_test, RNN_pred)
print('Accuracy: %f' % RNN_accuracy)
print()
# precision tp / (tp + fp)
RNN_precision = precision_score(y_test, RNN_pred, average='weighted')
print('Precision: %f' % RNN_precision)
print()
# recall: tp / (tp + fn)
RNN_recall = recall_score(y_test, RNN_pred, average='weighted')
print('Recall: %f' % RNN_recall)
print()
# f1: 2 tp / (2 tp + fp + fn)
RNN_f1 = f1_score(y_test, RNN_pred, average='weighted')
print('F1 score: %f' % RNN_f1)
print()
print()
print(classification_report(y_test, RNN_pred))
```
